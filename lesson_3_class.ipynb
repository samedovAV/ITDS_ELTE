```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.cos(x))
plt.plot(x, np.tan(x))
plt.plot(x. np.tanh(x))
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.exp(x))
```

```python
x = np.linspace(0, 10)

plt.plot(x, x)
plt.plot(x, 2+x+2, '--')
plt.plot(x, 3+x+6, '+')
```

```python
plt.plot(x, x)
plt.plot(x, 2+x+2, 'solid')
plt.plot(x, 3+x+6, '+')
```

```python
plt.plot(x, np.sin(x))
plt.title('A sine')
```
